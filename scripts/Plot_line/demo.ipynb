```python
SFC_TMP = ds['SFC_TMP'][:,0,:,:].mean(dim=['y','x'])
RH      = ds['RH'][:,0,:,:].mean(dim=['y','x'])
WSPD10  = ds['WSPD10'][:,0,:,:].mean(dim=['y','x'])
SOL_RAD = ds['SOL_RAD'][:,0,:,:].mean(dim=['y','x'])

SFC_TMP.shape
```

```python
from ydata_profiling import ProfileReport
import pandas as pd
```

```python
df_SFC_TMP = SFC_TMP.to_dataframe().drop(columns=['level'])
df_RH = RH.to_dataframe().drop(columns=['level'])
df_WSPD10 = WSPD10.to_dataframe().drop(columns=['level'])
df_SOL_RAD = SOL_RAD.to_dataframe().drop(columns=['level'])

df = pd.concat([df_SFC_TMP, df_RH, df_WSPD10, df_SOL_RAD], axis=1)
df
```

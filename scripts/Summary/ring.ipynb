```python
data = {
    'year': [24.16, 71.69, 100-24.16-71.69],
    'season': [61.55, 30.41, 100-61.55-30.41],
    'week': [85.8, 13.3, 100-85.8-13.3]
}

colors = ['#78add2', '#ffb26e', '#d8d8d8']

fig, ax = plt.subplots(figsize=(6, 6), dpi=300)

wedges = []
autotexts = []

for i, (label, values) in enumerate(data.items()):
    wedge = ax.pie(values, radius=1-0.2*i, colors=colors, wedgeprops=dict(width=0.15), autopct='%1.2f%%',
                   pctdistance=0.95, startangle=90, counterclock=False)[0]
    wedges.extend(wedge)
    autotexts.extend(plt.gca().texts)

plt.legend(['$\Delta$MET', '$\Delta$EMIS', 'Other'], loc=1)
ax.axis('equal')
plt.setp(autotexts, size=14, weight="bold")

plt.tight_layout()
plt.show()
```

```python
data = {
    'Year': [24.16, 71.69, 100-24.16-71.69],
    'Season': [61.55, 30.41, 100-61.55-30.41],
    'Week': [85.8, 13.26, 100-85.8-13.26]
}

colors = ['#78add2', '#ffb26e50', '#d8d8d850']

sizes = [0.4, 0.7, 1.0]

fig, ax = plt.subplots(figsize=(6, 6), dpi=300)

wedgeprops = {'width': 0.2, 'edgecolor': 'k', 'lw': 3}
otherprops = dict(wedgeprops=wedgeprops, startangle=90, colors=colors,
                  counterclock=False)

for key, values in data.items():
    radius = sizes.pop(0)
    patches, _ = ax.pie(values, radius=radius, **otherprops)
    patches[1].set_zorder(0)
    patches[1].set_edgecolor('white')
    patches[2].set_zorder(0)
    patches[2].set_edgecolor('white')

for i, (key, _) in enumerate(reversed(data.items())):
    ax.text(-0.05, 0.9-0.3*i, f'{key}:{data[key][0]:.1f}%', ha='right', va='center', fontsize=16, fontweight='bold',
            bbox=dict(facecolor='#8cc33480', edgecolor='none', boxstyle='round,pad=0.3'))

plt.legend(['$\Delta$MET', '$\Delta$EMIS', 'Other'], loc=1)
ax.axis('equal')

plt.tight_layout()
plt.show()
```
